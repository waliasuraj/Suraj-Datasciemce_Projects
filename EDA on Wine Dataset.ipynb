```python
# Visualizing the Data
# A visualization is a powerful tool in EDA that helps in understanding 
# the patterns and relationships in the data. We will use several plots to visualize the data, 
#such as histograms, scatter plots, and bar plots.
```

```python
#graph all the data set - just for looking
wine.plot(figsize=(15,7))
```

```python
sns.pairplot(wine)
plt.show()
```

```python
sns.boxplot(x=wine["alcohol"])
```

```python
# Histogram of wine quality
plt.hist(wine["quality"], color="green")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.title("Histogram of Wine Quality")
plt.show()
```

```python
wine.hist()
plt.show()
```

```python
# Scatter plot of alcohol and quality
plt.scatter(wine["alcohol"], wine["quality"])
plt.xlabel("Alcohol Content")
plt.ylabel("Quality")
plt.title("Scatter Plot of Alcohol Content and Quality")
plt.show()
```

```python
# Bar plot of wine pH and quality
sns.barplot(x="quality",y="pH", data=wine)
plt.xlabel("quality")
plt.ylabel("pH")
plt.title("Bar Plot of pH and Quality")
plt.show()
```

```python
# Bar plot of alcohol and quality
sns.barplot(x="quality",y="alcohol", data=wine)
plt.xlabel("quality")
plt.ylabel("alcohol")
plt.title("Bar Plot of alcohol and Quality")
plt.show()
```

```python
#We can observe that average Alcohol increasing with the increase in quality 
```

```python
# Correlation Analysis
# Here perform a correlation analysis to understand the relationships between the attributes in the data.
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(wine.corr(),annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
# now we see the effect of the elements on the quality
```

```python
wine.columns 
```

```python
# let see effect some of elements on the quality - details
plt.figure(figsize=(20,7))
sns.lineplot(data=wine, x="quality",y="volatile acidity",label="Volatile Acidity")
sns.lineplot(data=wine, x="quality",y="citric acid",label="Citric Acid")
sns.lineplot(data=wine, x="quality",y="chlorides",label="chlorides")
sns.lineplot(data=wine, x="quality",y="free sulfur dioxide",label="free sulfur dioxide")
sns.lineplot(data=wine, x="quality",y="total sulfur dioxide",label="total sulfur dioxide")
sns.lineplot(data=wine, x="quality",y="pH",label="PH")
sns.lineplot(data=wine, x="quality",y="residual sugar",label="residual sugar")
sns.lineplot(data=wine, x="quality",y="sulphates",label="Sulphates")
plt.ylabel("Quantity")
plt.title("Impact on quality")
plt.legend()
plt.show()
```

```python
# effect the Alcohol in the quality

plt.figure(figsize=(15,7))
sns.lineplot(data=wine, x="quality",y="alcohol",label="alcohol",color="green")
```

```python
# effect the total sulfur dioxide in the quality
plt.figure(figsize=(15,7))
sns.lineplot(data=wine, x="quality",y="total sulfur dioxide",label="total sulfur dioxide")
```

```python
# effect the free sulfur dioxide in the quality

plt.figure(figsize=(15,7))
sns.lineplot(data=wine, x="quality",y="free sulfur dioxide",label="free sulfur dioxide",color="y")
```
